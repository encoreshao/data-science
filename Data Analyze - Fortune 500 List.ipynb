```python
fortune_500_companies = 'resources/Fortune500Companies.xlsx'
```

```python
import pandas as pd
```

# Reading data from Excel

```python
df = pd.read_excel(fortune_500_companies)
```

```python
# Filtering Data

# Select Employees and Profits
dfEmployees = df.loc[(df['Number of Employees'] == 2200000) & (df['Profits'] == 14881)]
```

```python
dfEmployees
```

```python
# Select Employees

dfMoreThanEmployees = df.loc[(df['Number of Employees'] > 137000)]
```

```python
dfMoreThanEmployees
```

```python
change_in_rank = 'Change in Rank'
```

# Line Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

line_change_in_rank = dfMoreThanEmployees[change_in_rank].value_counts()
line_change_in_rank.plot()

plt.title(change_in_rank)
plt.ylabel('Rank')
plt.xlabel('Change')
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

# Pie Chart

```python
pie_change_in_rank = pd.DataFrame(dfMoreThanEmployees[change_in_rank].value_counts())
pie_change_in_rank.columns=[change_in_rank]
pie_change_in_rank.plot(kind='pie', y=change_in_rank)
```

# Bar Chart

```python
bar_change_in_rank = dfMoreThanEmployees[change_in_rank].value_counts()
bar_change_in_rank.sort_values().plot(kind='barh')
plt.title(change_in_rank)
plt.ylabel('Rank')
plt.xlabel('Change')
```
